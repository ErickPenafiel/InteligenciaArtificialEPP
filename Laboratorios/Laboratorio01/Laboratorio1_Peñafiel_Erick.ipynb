```python
import pprint as p
import random

def crearTablero(filas, columnas, lista):
  for i in range(filas):
    lista.append([])
    for j in range(columnas):
      if i==0 or i==filas-1:
        lista[i].append('#')
      elif j == 0 or j==columnas-1:
        lista[i].append('#')
      elif i%2 == 0 and j%2 == 0:
        lista[i].append('#')
      else:
        lista[i].append(' ')
  return lista 
```

```python
def contarParedes(filas, columnas, listas):
  cont = 0
  for i in range(1, filas-1):
    for j in range(1, columnas-1):
      if lista[i][j] == '#':
        cont += 1

  return cont
```

```python
def obstaculos(filas, columnas, listas, obst):
  numParedes = contarParedes(filas, columnas, listas)
  print(numParedes)
  i=1
  if numParedes < obst:
    total = obst - numParedes

    while i <= total:
      x = random.randint(1, filas-2)
      y = random.randint(1, columnas-1)
      if lista[x][y] == ' ':
        lista[x][y] = '#'
        i+=1
  else:
    total = numParedes - obst

    while i <= total:
      x = random.randint(1, filas-2)
      y = random.randint(1, columnas-1)
      if lista[x][y] == '#':
        lista[x][y] = ' '
        i+=1
  return lista
```

```python
def jugador(filas, columnas, lista):
  cond = True
  while cond:
    x = random.randint(1, filas-2)
    y = random.randint(1, columnas-2)
    if lista[x][y] == ' ':
      lista[x][y] = 'G'
      cond = False

  return lista
```

```python
def mover(fila, columna, lista):
  for i in range(fila):
    for j in range(columna):
      if lista[i][j] == 'G':
        posX = j
        posY = i
  
  print("X: ", posX)
  print("Y: ", posY)

  mov = True
  while mov:
    movimiento = random.randint(1, 4)
    #print(movimiento)

    if movimiento == 1:
      if lista[posY-1][posX] == ' ':
        lista[posY][posX] = ' '
        lista[posY-1][posX] = 'G'
        mov = False
        #print("Entro 1")
    elif movimiento == 2:
      if lista[posY][posX+1] == ' ':
        lista[posY][posX] = ' '
        lista[posY][posX+1] = 'G'
        mov = False
        #print("Entro 2")
    elif movimiento == 3:
      if lista[posY+1][posX] == ' ':
        lista[posY][posX] = ' '
        lista[posY+1][posX] = 'G'
        mov = False
        #print("Entro 3")
    else:
      if lista[posY][posX-1] == ' ':
        lista[posY][posX] = ' '
        lista[posY][posX-1] = 'G'
        mov = False
        #print("Entro 4")
  return lista
```

```python
filas = int(input('Ingrese cantidad de filas: '))
columnas = int(input('Ingrese la cantidad de columnas: '))
obst = int(input('Ingrese cantidad de obstaculos: '))
lista = []
```

```python
lista = crearTablero(filas, columnas, lista)
p.pprint(lista)
```

```python
obstaculos(filas, columnas, lista, obst)
p.pprint(lista)
```

```python
jugador(filas, columnas, lista)
p.pprint(lista)
```
